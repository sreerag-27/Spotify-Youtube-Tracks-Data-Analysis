### Q3. Artist 'The Weeknd'  With highest Streams on Spotify and Youtube

```python
spotify_youtube_df[['Artist','Track','Stream']].sort_values('Stream',ascending = False)
```

```python
the_weeknd_spotify_df = spotify_youtube_df[['Artist','Track','Stream']][spotify_youtube_df.Artist == 'The Weeknd']
the_weeknd_spotify_df
```

```python
plt.figure(figsize=(30,4))

plt.xticks(rotation = 30)

sns.set(font_scale = 1.7)

sns.lineplot(data = the_weeknd_spotify_df , x = 'Track', y = 'Stream');
```

__In Spotify Streams Of Artist `The Weeknd` Track `Blinding Lights` has the Peak Highest with more than 3 billion Streams__

```python
the_weeknd_youtube_df = spotify_youtube_df[['Artist','Track','Views']][spotify_youtube_df.Artist == 'The Weeknd']
the_weeknd_youtube_df 
```

```python
plt.figure(figsize=(30,4))
plt.xticks(rotation = 30)
sns.lineplot(data = the_weeknd_youtube_df , x = 'Track',y = 'Views');
```

__On Youtube Of Artist `The Weeknd` there are almost 2 times peak highest views on track first is `Starboy` with more than 2 billion streams and another is `The Hills` Track with more than 1.5 billion views.__
